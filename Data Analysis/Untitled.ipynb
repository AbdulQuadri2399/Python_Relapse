```python
import csv
albumlist = []
with open("albumlist.csv", "r") as csvFile:
    csvDict = csv.DictReader(csvFile)
    for row in csvDict:
        albumlist.append(row)
```

```python
yearlist = [row["Year"] for row in albumlist]
print(yearlist)
```

```python
def is_valid_year(year):
    try:
        value = int(year)
    except ValueError:
        return False
    else:
        return value > 1400
```

```python
validReleaseYears = [int(row["Year"]) for row in albumlist if is_valid_year(row["Year"])]
```

```python
minYear = min(validReleaseYears)
print(minYear)
```

```python
sortedAlbumList = sorted(albumlist, key = lambda x: x["Artist"])
```

```python
sortedYear = sorted(validReleaseYears)
```

```python
minFunc = lambda x: min(x)
print(minFunc(sortedYear))
```

```python
maxFunc = lambda x: str(max(x))
```

```python
print(maxFunc(sortedYear))
```

```python
valid_albums = [row for row in albumlist if is_valid_year(row["Year"])]
int_years_album = [int(row["Year"]) for row in valid_albums]
```

```python
for row in valid_albums:
    for year in row:
        print(type(year))
```

```python
album_max = max(valid_albums, key = lambda x: x["Year"])
print(album_max["Album"], album_max["Artist"], album_max["Year"], type(album_max["Year"]))
```

```python
max_year_album = [row for row in valid_albums if row["Year"] == str(max(int_years_album))]
print(max_year_album)
```
